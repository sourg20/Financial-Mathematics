```python
import numpy as np
import matplotlib.pyplot as plt

# Define parameters
St = np.linspace(100, 170, 100)  # Possible values of St
K1 = 130  # Strike price of put option
K2 = 140  # Strike price of call option
P = 5  # Assumed price of put option
C = 10  # Assumed price of call option

# Calculate payoff
payoff_put = np.maximum(K1 - St, 0)
payoff_call = np.maximum(St - K2, 0)
total_payoff = payoff_put - payoff_call

# Calculate profit
profit_put = payoff_put - P
profit_call = payoff_call - C
total_profit = profit_put + profit_call

# Plot payoff
plt.figure(figsize=(12, 6))
plt.plot(St, total_payoff, label='Total Payoff')
plt.plot(St, payoff_put, label='Put Payoff')
plt.plot(St, payoff_call, label='Call Payoff')
plt.title('Payoff Diagram')
plt.xlabel('St')
plt.ylabel('Payoff')
plt.legend()
plt.grid()
plt.show()

# Plot profit
plt.figure(figsize=(12, 6))
plt.plot(St, total_profit, label='Total Profit')
plt.plot(St, profit_put, label='Put Profit')
plt.plot(St, profit_call, label='Call Profit')
plt.title('Profit Diagram')
plt.xlabel('St')
plt.ylabel('Profit')
plt.legend()
plt.grid()
plt.show()
```
